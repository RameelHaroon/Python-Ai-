```python
class Node:
    def __init__(self,State,parent=None,cost=0):
        self.state = State[:]
        self.parent = parent
        self.cost = cost
        self.children = []
```

```python
from collections import deque

def TreeSearch(S, G):
    queue = deque()
    queue.append(S)
    Explored = []

    while True:
        if len(queue) > 0:
            node = queue.popleft()
            Expand(node)

            if node.state == G:

                print('Goal State found')
                print(f'Total Moves = {node.cost}')
                PrintGame(node)
               
                break
            for x in node.children:
                if x.state not in Explored:
                    Explored.append(x.state)
                    x.parent = node
                    x.cost = node.cost + 1
                    queue.append(x)
        else:
            print("Goal State Not Found!")
            break


def Expand(a):
    BlankPosition = a.state.index('blank')

    if BlankPosition == 0 or BlankPosition == 1 or BlankPosition == 2 or BlankPosition == 3 or BlankPosition == 4 or BlankPosition == 5:
        a.children.append(MoveDown(a.state, BlankPosition))
    if BlankPosition == 3 or BlankPosition == 4 or BlankPosition == 5 or BlankPosition == 6 or BlankPosition == 7 or BlankPosition == 8:
        a.children.append(MoveUp(a.state, BlankPosition))
    if BlankPosition == 1 or BlankPosition == 2 or BlankPosition == 4 or BlankPosition == 5 or BlankPosition == 7 or BlankPosition == 8:
        a.children.append(MoveLeft(a.state, BlankPosition))
    if BlankPosition == 0 or BlankPosition == 1 or BlankPosition == 3 or BlankPosition == 4 or BlankPosition == 6 or BlankPosition == 7:
        a.children.append(MoveRight(a.state, BlankPosition))


def MoveDown(state, BlankPosition):
    temp = state[:]
    temp[BlankPosition] = temp[BlankPosition + 3]
    temp[BlankPosition + 3] = 'blank'

    return Node(temp)


def MoveUp(state, BlankPosition):
    temp = state[:]
    temp[BlankPosition] = temp[BlankPosition - 3]
    temp[BlankPosition - 3] = 'blank'

    return Node(temp)


def MoveLeft(state, BlankPosition):
    temp = state[:]
    temp[BlankPosition] = temp[BlankPosition - 1]
    temp[BlankPosition - 1] = 'blank'

    return Node(temp)


def MoveRight(state, BlankPosition):
    temp = state[:]
    temp[BlankPosition] = temp[BlankPosition + 1]
    temp[BlankPosition + 1] = 'blank'

    return Node(temp)

def PrintGame(node):
            temp = node 
            stack = []
            while temp is not None:
                stack.append(temp.state)
                temp = temp.parent
                
            for x in range(len(stack)):
                Move = stack.pop()
                i = 1
                for m in Move:
                    if i % 3 == 0:
                        if m == 'blank':
                            print('-') 
                        else:
                            print(m)
                    else:
                        if m == 'blank':
                            print('-',end=" ") 
                        else:
                            print(m,end=" ")
                    i = i + 1
                print('  |')
                print('  ↓')
```

```python
Source = Node([3,4,1,2,5,6,7,8,'blank'])
Goal = [1,2,3,4,5,6,7,8,'blank']
TreeSearch(Source,Goal)
```
