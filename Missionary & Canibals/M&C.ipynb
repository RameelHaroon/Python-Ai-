```python
class Node:
    def __init__(self,State,parent=None,cost=0):
        self.state = State[:]
        self.parent = parent
        self.cost = cost
        self.children = []
```

```python
from collections import deque

def TreeSearch(S, G):
    queue = deque()
    queue.append(S)
    Explored = [S.state]

    while True:
        if len(queue) > 0:
            node = queue.popleft()

            if (LegalState(node)):
                Expand(node)

                if node.state == G:
                    print('Goal State found')
                    print(f'Total Moves = {node.cost}')


                    temp = node
                    stack = []
                    while temp is not None:
                        stack.append(temp.state)
                        temp = temp.parent

                    while len(stack) > 0:
                        x = stack.pop()
                        print(x)
                    break
                for x in node.children:
                    if x.state not in Explored:

                        x.parent = node
                        x.cost = node.cost + 1
                        if not LegalState(x):
                            Explored.append(x.state)
                        if LegalState(x):
                            queue.append(x)
        else:

            print("Goal State Not Found!")
            break


def Expand(a):
    boat = a.state[2]

    if boat == 1:

        a.children.append(OneMis(a.state))
        a.children.append(OneMis(a.state))
        a.children.append(OneCan(a.state))
        a.children.append(TwoMis(a.state))
        a.children.append(TwoCan(a.state))
        a.children.append(OneMisOneCan(a.state))
    else:
        a.children.append(OneMisBack(a.state))
        #a.children.append(TwoMisBack(a.state))
        a.children.append(OneMisOneCanBack(a.state))
        #a.children.append(TwoCanBack(a.state))
        a.children.append(OneCanBack(a.state))


def LegalState(node):
    if node.state[0] < 0 or node.state[1] < 0 or node.state[0] > 3 or node.state[1] > 3:
        return False
    if node.state[1] > node.state[0] and node.state[0] > 0:
        return False
    if node.state[1] < node.state[0] and node.state[0] < 3:
        return False
    return True
def OneMis(a):
    temp = a[:]
    temp[0] = temp[0] - 1
    temp[2] = 0
    return Node(temp)


def OneCan(a):
    temp = a[:]
    temp[1] = temp[1] - 1
    temp[2] = 0
    return Node(temp)


def TwoMis(a):
    temp = a[:]
    temp[0] = temp[0] - 2
    temp[2] = 0
    return Node(temp)


def TwoCan(a):
    temp = a[:]
    temp[1] = temp[1] - 2
    temp[2] = 0
    return Node(temp)


def OneMisOneCan(a):
    temp = a[:]
    temp[0] = temp[0] - 1
    temp[1] = temp[1] - 1
    temp[2] = 0
    return Node(temp)


def OneMisBack(a):
    temp = a[:]
    temp[0] = temp[0] + 1
    temp[2] = 1
    return Node(temp)


def OneCanBack(a):
    temp = a[:]
    temp[1] = temp[1] + 1
    temp[2] = 1
    return Node(temp)

def TwoMisBack(a):
    temp = a[:]
    temp[0] = temp[0] + 2
    temp[2] = 1
    return Node(temp)


def TwoCanBack(a):
    temp = a[:]
    temp[1] = temp[1] + 2
    temp[2] = 1
    return Node(temp)


def OneMisOneCanBack(a):
    temp = a[:]
    temp[0] = temp[0] + 1
    temp[1] = temp[1] + 1
    temp[2] = 1
    return Node(temp)
```

```python
Source = Node([3,3,1])
Goal = [0,0,0]
TreeSearch(Source,Goal)
```
